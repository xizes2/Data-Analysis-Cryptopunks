```python
import pandas as pd
import numpy as np
```

## TRANSACTIONS

```python
trans_df = pd.read_csv('./CSV/punks_transactions.csv', index_col=0)
trans_df.info()
```

```python
trans_df['Amount'] = trans_df['Amount'].map(lambda x: str(x)[:-1])
trans_df['Trans_id'] = trans_df['Date'].map(lambda x: str(x)[-8:])
trans_df['Date'] = trans_df['Date'].map(lambda x: str(x)[:-8])
trans_df['Punk_id'] = trans_df['Punk_id'].map(lambda x: str(x)[12:])
trans_df
```

```python
trans_df['Amount'] = trans_df['Amount'].replace({'na':'0'})
trans_df['Amount'] = trans_df['Amount'].str.replace(',','').astype('float')
trans_df['Date'] = pd.to_datetime(trans_df['Date'])
trans_df.info()
```

```python
trans_df.set_index('Trans_id', inplace=True)
trans_df
```

```python
trans_df.to_csv('./CSV/transactions_ready.csv')
```

## PUNkS

```python
punks_df = pd.read_csv('./CSV/punks_list.csv', index_col=0, thousands=",")
punks_df.info()
```

```python
punks_df = punks_df.rename(columns={'Punk':'Punk_id'})
punks_df['Skin'] = punks_df['Skin'].map(lambda x: str(x)[:-8])
punks_df['Punk_id'] = punks_df['Punk_id'].map(lambda x: str(x)[12:])
punks_df
```

```python
punks_df.info()
```

```python
punks_df.set_index('Punk_id', inplace=True)
punks_df
```

```python
punks_df.to_csv('./CSV/punks_ready.csv')
```

## TRAITS BY PUNK

```python
punks_traits = punks_df['Traits'].copy()
punks_traits
```

```python
traits_by_punk = []
for i in punks_traits.index:
    atts = punks_traits.loc[i]
    atts = atts[1:-1].split(',')
    for att in list(atts):
        ident = [i, att]
        traits_by_punk.append(ident)
```

```python
traits_by_punk_df = pd.DataFrame.from_records(traits_by_punk, columns=['punk_id', 'Trait'])
traits_by_punk_df['Trait'] = traits_by_punk_df['Trait'].str.strip()  
traits_by_punk_df.set_index('punk_id', inplace=True)
traits_by_punk_df
```

```python
traits_by_punk_df.to_csv('./CSV/traits_by_punk_ready.csv')
```

```python
traits_count_df = pd.DataFrame(traits_by_punk_df['Trait'].value_counts())
traits_count_df
```

```python
traits_count_df.to_csv('./CSV/traits_counts_ready.csv')
```
