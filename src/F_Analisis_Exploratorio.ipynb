**Sobre Outliers x CryptoArte:**

Nosotros entendemos que el concepto de outliers aplicado a los bienes digitales, y en especial los cryptopunks, no debe ser aplicado de la misma manera como en muchas otras análisis de datos (quizá la mayoría). Esto se debe porque el mercado de arte en si ya carga una variable especulativa, (y hasta cierto punto de borbuja), con grandes variaciones de precios. A eso se suma una de las características más fuertes del mercado crypto en general, la especulación (otra vez), además de otras razones como el hecho de que aún es un mercado nuevo, la cantidad de participantes baja, su uso para actividades ilícitas, menor transparencia y control, entre otros.

Por esto consideramos que algunas oportunidades están justamente en estas características lo que justificaria el mantenimiento de algunos datos que en otras análisis seguramente serian considerados outliers y consequentemente removidos del dataset.
