```python
# Import docx library to handle word documents
#!pip install python-docx
from docx import Document # document object
from docx.enum.text import WD_ALIGN_PARAGRAPH, WD_BREAK # paragraph alignment and page break
from docx.shared import Pt
```

```python
# Create document object
document = Document()

# Create 1st header
header_1 = document.add_heading("Madrid's Stock Exchange webpage Scrapping", 0)
# Center header
header_1.alignment = WD_ALIGN_PARAGRAPH.CENTER

# Create 2nd header
document.add_heading("Description", 1)

# Create 1st paragraph
p = document.add_paragraph("")
# Create run inside paragraph 
# Obs.: In order to make changes to the font formatting we have to put the text on the 'run' level, not 'paragraph'
p.add_run("The purpose of this exercise is to create this word file and describe/explain what we've done on the 1st exercise, when we scraped the Madrid Stock Exchange webpage.")
# Change font size on the specific 'run' level we want
document.paragraphs[2].runs[0].font.size = Pt(13)
# Change font name on the specific 'run' level we want
document.paragraphs[2].runs[0].font.name = 'Verdana'

# Same steps as before...
p2 = document.add_paragraph("")
p2.add_run("A table on the initial page, showing the first 5 companies on IBEX35 index, was scraped using Selenium library.")
document.paragraphs[3].runs[0].font.size = Pt(13)
document.paragraphs[3].runs[0].font.name = 'Verdana'

# Same steps as before...
p3 = document.add_paragraph("")
p3.add_run("The columns of the dataframe created using the information scraped are: ")
document.paragraphs[4].runs[0].font.size = Pt(13)
document.paragraphs[4].runs[0].font.name = 'Verdana'

# Create bullet list containing columns names
document.add_paragraph("company name", style="List Bullet")
document.add_paragraph("last price", style="List Bullet")
document.add_paragraph("variation.", style="List Bullet")
# Format the text same way as before
document.paragraphs[5].runs[0].font.size = Pt(13)
document.paragraphs[5].runs[0].font.name = 'Verdana'
document.paragraphs[6].runs[0].font.size = Pt(13)
document.paragraphs[6].runs[0].font.name = 'Verdana'
document.paragraphs[7].runs[0].font.size = Pt(13)
document.paragraphs[7].runs[0].font.name = 'Verdana'


# Justify all the paragraphs
p.alignment = WD_ALIGN_PARAGRAPH.JUSTIFY
p2.alignment = WD_ALIGN_PARAGRAPH.JUSTIFY
p3.alignment = WD_ALIGN_PARAGRAPH.JUSTIFY

#Create blank paragraph to separate table from text
document.add_paragraph()

# Create table with the information scraped
table = document.add_table(rows=1, cols=df.shape[1])
hdr_cells = table.rows[0].cells
hdr_cells[0].text = df.columns[0]
hdr_cells[1].text = df.columns[1]
hdr_cells[2].text = df.columns[2]

# Feed table with the information on df
for a, b, c in zip(df['Nombre'], df['Último'], df['%Dif.']):
    row_cells = table.add_row().cells
    row_cells[0].text = a
    row_cells[1].text = b
    row_cells[2].text = c
    

# Save text with the format we want in a word file
document.save("test.docx")
```

```python
# Install and import scrapy

#conda install -c conda-forge scrapy
from scrapy.http import TextResponse
```

```python
# Select web to be scrapped
url = 'https://www.imdb.com/chart/moviemeter/?ref_=nv_mv_mpm'
```

```python
# Get url using 'requests' library
a = requests.get(url)
```

```python
a.url
```

```python
# web html in text format
a.text
```

```python
# Create 'response' object to be scraped
r = TextResponse(a.url, body=a.text, encoding='utf-8')
```

```python
# Feed the table on the page containing the information we want
table_movies = r.xpath("//tr")
```

```python
#table_movies
```

```python
# Create dictionary to receive the information we want
dict_movies = {'Movies':[], 'Ratings':[], 'Directors':[]}

# Loop through the table created and extract each info we want
for i, movie in enumerate(table_movies[1:]):
    
    # Jump first pass because it´s the header
    if i == 0:
        pass
    
    # Start collecting info
    # I used 'try, except' because some elements were missing
    else:
        # Movie Name 
        # Append movie name found on the xpath of each item to dictionary (I just had to use some string methods to isolate the text I wanted)
        try:
            dict_movies['Movies'].append(movie.xpath('//*[@id="main"]/div/span/div/div/div[3]/table/tbody/tr['+str(i)+']/td[2]/a').get().split('>')[1].replace('</a', ''))
        except:
            print('Movie: N/A')
            dict_movies['Movies'].append('N/A')
       
        # Movie Rating
        # Append rating following the same logic as before
        try:
            dict_movies['Ratings'].append(movie.xpath('//*[@id="main"]/div/span/div/div/div[3]/table/tbody/tr['+str(i)+']/td[3]/strong').get().split('>')[1].replace('</strong', ''))
        except:
            print('Rating: N/A')
            dict_movies['Ratings'].append('N/A')
            
        # Movie Director
        # Append director following the same logic as before
        try:
            dict_movies['Directors'].append(movie.xpath('//*[@id="main"]/div/span/div/div/div[3]/table/tbody/tr['+str(i)+']/td[2]/a').attrib['title'].split(',')[0].split('(')[0].strip())
        except:
            print('Director: N/A')
            dict_movies['Directors'].append('N/A')
```
